```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from keras.layers import Conv2D, MaxPool2D, AveragePooling2D, Input, BatchNormalization, MaxPooling2D, Activation, Flatten, Dense, Dropout
from keras.models import Sequential
from keras.utils import np_utils
from sklearn.metrics import classification_report
from imblearn.over_sampling import RandomOverSampler
from keras.preprocessing import image
import scipy
import os
import cv2
```

```python
data = pd.read_csv("./real5foldercsv.csv")
data.shape
```

```python
folder_train_data = data[data['folder'] != "folder1"]
folder_train_data.shape
```

```python
folder_val_data = data[data['folder'] == "folder1"]
folder_val_data.shape
```

```python
x_folder_train_data = folder_train_data['pixels']
y_folder_train_data = folder_train_data['emotion']
```

```python
x_folder_val_data = folder_val_data['pixels']
y_folder_val_data = folder_val_data['emotion']
```

```python
x_folder_val_data
```

```python
oversampler = RandomOverSampler(sampling_strategy='auto')
x_folder_train_data, y_folder_train_data = oversampler.fit_resample(x_folder_train_data.values.reshape(-1,1), y_folder_train_data)
print(x_folder_train_data.shape," ",y_folder_train_data.shape)
x_folder_val_data, y_folder_val_data = oversampler.fit_resample(x_folder_val_data.values.reshape(-1,1), y_folder_val_data)
print(x_folder_val_data.shape," ",y_folder_val_data.shape)
```

```python
x_folder_train_data = pd.Series(x_folder_train_data.flatten())
x_folder_val_data = pd.Series(x_folder_val_data.flatten())
```

```python
x_folder_train_data = np.array(list(map(str.split, x_folder_train_data)), np.float32)
x_folder_train_data/=255



y_folder_train_data = np.array(y_folder_train_data)
y_folder_train_data = y_folder_train_data.reshape(y_folder_train_data.shape[0], 1)
y_folder_train_data.shape
```

```python
x_folder_val_data = np.array(list(map(str.split, x_folder_val_data)), np.float32)
x_folder_val_data/=255
```

```python
x_folder_val_data.shape
```

```python
y_folder_train_data = np_utils.to_categorical(y_folder_train_data, 8)
y_folder_val_data = np_utils.to_categorical(y_folder_val_data, 8)
```

```python
model = keras.Sequential([
    Input(x_folder_train_data.shape[1]),
    Dense(2048, activation = 'relu'),
    BatchNormalization(axis = 1),
    Dense(1024, activation = 'relu'),
    BatchNormalization(axis = 1),
    Dense(512, activation = 'relu'),
    BatchNormalization(axis = 1),
    Dense(256, activation = 'relu'),
    BatchNormalization(axis = 1),
    Dense(8, activation = 'sigmoid')
])
 
model.summary()
```

```python
tf.keras.utils.plot_model(model, to_file = "model.png", show_shapes = True, show_dtype = True)
```

```python
adam = keras.optimizers.Adam(learning_rate=0.0001)
model.compile(optimizer=adam, loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x_folder_train_data, y_folder_train_data, validation_data = (x_folder_val_data,y_folder_val_data), epochs = 100)
```

```python
print("Accuracy of our model on validation data : " , model.evaluate(x_folder_val_data,y_folder_val_data)[1]*100 , "%")
```
