```python
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor('./shape_predictor_68_face_landmarks.dat')

for folder in os.listdir(main_path_image):
   
        
    
    for file in os.listdir(main_path_image+"/"+folder):
        image = cv2.imread(main_path_image+"/"+folder+"/"+file)
        image = cv2.resize(image,(512,512))
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        rects = detector(gray, 1)
        if not rects :
            print("not_pass :",count_not_pass)
            count_not_pass = count_not_pass +1
            
        else:
            #0=anger,1=contempt,2=disgust,3=fear,4=happiness,5=neutral,6=sadness,7=surprise
            file_pass.append(file)
            if count_file == 0 :
                emotion_featured.append(0)
                count_anger = count_anger +1
            elif count_file == 1 :
                emotion_featured.append(1)
                count_contempt = count_contempt +1
            elif count_file == 2 :
                emotion_featured.append(2)
                count_disgust = count_disgust +1
            elif count_file == 3 :
                emotion_featured.append(3)
                count_fear = count_fear+1
                
            elif count_file == 4 :
                emotion_featured.append(4)
                count_happiness = count_happiness +1
            
            elif count_file == 5 :
                emotion_featured.append(5)
                count_neutral = count_neutral +1
            elif count_file == 6 :
                emotion_featured.append(6)
                count_sadness = count_sadness +1
            elif count_file == 7 :
                emotion_featured.append(7)
                count_surprise = count_surprise +1
            print("pass :",count_pass)
            count_pass = count_pass +1
            
            for rect in rects:
                shape = predictor(gray, rect)
                shape_np = np.zeros((68, 2), dtype="int")
                for i in range(0, 68):
                    shape_np[i] = (shape.part(i).x, shape.part(i).y)
                shape = shape_np
                x_target=0
            
                for i, (x, y) in enumerate(shape):
                    if i == 30:
                        print(x,y)
                        cv2.circle(image, (x, y), 1, (0, 255, 0), 30)
                        
                    elif i <30 and i> 16 :
                        cv2.line(image, (214,262), (x,y),(102, 255, 255), 1)
                        cv2.circle(image, (x, y), 1, (255, 0, 0), 10)
                    else:
                        cv2.line(image, (214,262), (x,y),(102, 255, 255), 1)
                        cv2.circle(image, (x, y), 1, (0, 0, 255), 10)
            for i, (x, y) in enumerate(shape):
                if i ==30:
                    x_main = x
                    y_main = y
            distance = []
            for i, (x, y) in enumerate(shape):
               
                
                if i >=0 and i <68 and i != 30:
                    result = spatial.distance.cosine((x,y),(x_main,y_main))
                    distance.append(result)
            index = []
            for x in range(0,68):
                if x != 30:
                    index.append(x)
        
            dict = {'Path':file ,'index':index,'distance':distance,}
            shutil.copyfile(main_path_image+"/"+folder+"/"+file,checked_feature_path+"/"+folder+"/"+file)
            path_output = r'C:\Users\ANFIELD\Desktop\tu\cn240\ann\score_csv'
            df = pd.DataFrame(dict)
            df.to_csv(path_output+'/'+folder+"/"+file+'_csv'+'.csv', mode='w+', header="test",index=False) 
            
            
    count_file = count_file +1
    
dict1 = {'Path':file_pass ,'emotion':emotion_featured,}
path_output_csvscore = r'C:\Users\ANFIELD\Desktop\tu\cn240\ann\csv_landmark'
df1 = pd.DataFrame(dict1)
df1.to_csv(path_output_csvscore+'/'+"new_score_featured.csv", mode='w+', header="test",index=False) 
        
        
```
